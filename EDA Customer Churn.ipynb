```python
# Converting the Value of the column Senior Citizen 1 ,0 to yes and no to amke it easier to understannd by creation a defination 
def conv(value):
    if value == 1:
        return "YES"
    else:
        return "NO"
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
plt.figure(figsize = (3,3))
group_by = df.groupby("Churn").agg({'Churn' :"count"})
plt.pie(group_by['Churn'], labels = group_by.index, autopct = "%1.2f%%")
plt.title( " Percentage Of Churned Customers ", fontsize = 12 )
plt.show()
```

```python
# From the abve Pie Chat we can conclude that nearly 27% of the customers have broke the relationship with us that means they have churned out .
# Now let us find the reason behind it .

plt.figure(figsize = (4,4))
sns.countplot(x = 'gender', data = df , hue = "Churn")
plt.show()
```

```python
plt.figure(figsize = (3,3))
ax = sns.countplot(x = 'SeniorCitizen', data = df , hue = "Churn", palette=["red", "grey"])
for container in ax.containers:
    ax.bar_label(container, padding = 10)
ax.set_title(" Count of Churned By Senior Citizen ", pad = 25)
plt.show()
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Plot
fig, ax = plt.subplots(figsize=(4, 4))  # Adjust figsize for better visualization

# Plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['skyblue', 'lightgreen'])  # Customize colors if desired

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor = (1.4,0.9))  # Customize legend location

plt.show()
```

```python
# From the above graph we came to know that % of senior Citizen have churned more 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Warm color pairs: (Churn=No, Churn=Yes)
color_pairs = [
    ('#FFB347', '#FFDAB9'),  # Mango / Light Peach
    ('#FF6961', '#FFB6B9'),  # Warm Red / Rose
    ('#F4A460', '#FFE4B5'),  # Sandy / Light Butter
    ('#FF7F50', '#FFDAB9'),  # Coral / Peach
    ('#E97451', '#F5CBA7'),  # Pumpkin / Apricot
    ('#D2691E', '#F4A460'),  # Choco / Sandy
    ('#CD5C5C', '#F08080'),  # Indian Red / Light Coral
    ('#FFA07A', '#FFEBCD'),  # Light Salmon / Cream
    ('#DAA520', '#FFE066')   # Goldenrod / Lemon Cream
]

columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(columns):
    churn_palette = {
        'No': color_pairs[i][0],
        'Yes': color_pairs[i][1]
    }

    # Create countplot
    sns.countplot(
        x=col,
        data=df,
        ax=axes[i],
        hue="Churn",
        palette=churn_palette
    )

    # Set titles and labels
    axes[i].set_title(f'Churn vs {col}', fontsize=20, fontweight='bold', pad=35)
    axes[i].set_xlabel(col, fontsize=20)
    axes[i].set_ylabel('Count', fontsize=20)

    # Add value labels
    for container in axes[i].containers:
        axes[i].bar_label(container, padding=3, fontsize=20)

    axes[i].legend(title='Churn')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle("Customer Churn Analysis by Services", fontsize=18, fontweight='bold', y=1.02)
plt.subplots_adjust(top=0.98)
plt.tight_layout()
plt.show()
```
